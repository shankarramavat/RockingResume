```python
from PyPDF2 import PdfReader
import os
import re
import csv
information = {}
for category in os.listdir("./data/data/data"):
    file_content = {}
    for file in os.listdir(f'./data/data/data/{category}'):
        all_content = {}
        reader = PdfReader(f'./data/data/data/{category}/{file}')
        number_of_pages = len(reader.pages)
        content = ""
        for i in range(0, number_of_pages):
            page = reader.pages[i]
            content += page.extract_text()
        pattern = r'Skills(.*?)(?:education|experience|work|professional|accomplishments|qualifications|relevant|$)'
        skills = re.search(pattern, content, re.DOTALL)
        extract_skills = ""
        if skills:
            extract_skills = skills.group(1)
            
        pattern2 = r'Education(.*?)(?:experience|skills|summary|work|professional|accomplishments|qualifications|relevant|publications|$)'
        education = re.search(pattern2, content, re.IGNORECASE | re.DOTALL)
        extract_education = ""
        if education:
            extract_education = education.group(1)
            
        pattern3 = r'^(.*?)(?:\bsummary\b|\brxecutive\b|\bskills\b|\bhighlights\b|\bexperience\b|\bcore\b|\bProfessional\b|\bcareer\b|\bprofile\b|\bposition\b|\bexecutive\b|\bobjective\b|\baccomplishments\b|\b$\b)'
        role = re.search(pattern3, content, re.IGNORECASE | re.DOTALL)
        extract_role = ""
        if role:
            extract_role = role.group(1)
        all_content["skills"] = extract_skills
        all_content["education"] = extract_education
        all_content["role"] = extract_role
        file_content[file] = all_content
    with open(f'extracted_files/{category}.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(["FileNumber", "Role", "Skills", "Education"])
        for keys in file_content:
            writer.writerow([keys, file_content[keys]["role"], file_content[keys]["skills"], file_content[keys]["education"]])
```
